```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, LSTM, GRU, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping

# Đảm bảo tính nhất quán dữ liệu khi tái chạy hệ thống
np.random.seed(42)
tf.random.set_seed(42)

# ==========================================
# 1. ĐỌC VÀ TIỀN XỬ LÝ DỮ LIỆU ĐĂNG KÝ MẪU (1/50)
# ==========================================
CSV_FILE_PATH = "/kaggle/input/datasets/thebluestar/my-dataflow/traffic.csv" 
SUB_SAMPLE_SIZE = 60710  # Cấu hình chính xác tỷ lệ 1/50 dữ liệu gốc (~60.710 dòng)

if not os.path.exists(CSV_FILE_PATH):
    print(f"[CẢNH BÁO] Không tìm thấy file {CSV_FILE_PATH} ở thư mục hiện tại.")
    print("Mẹo trên Kaggle: Hãy chắc chắn bạn đã tích hợp dữ liệu vào mục Input và cấu hình đúng đường dẫn.")
else:
    print(f"--> [HỆ THỐNG] Khởi tạo tiến trình trích xuất dữ liệu hệ thống...")
    df = pd.read_csv(CSV_FILE_PATH, nrows=SUB_SAMPLE_SIZE)

    # Chọn lọc các thuộc tính số học
    features = ['flow', 'occupy', 'speed']
    data = df[features].values

    # Chuẩn hóa dữ liệu về vùng toán học [0, 1]
    scaler = MinMaxScaler()
    data_scaled = scaler.fit_transform(data)

    # Cấu hình chuỗi thời gian quay lui (Dùng 12 bước quá khứ dự đoán 1 bước kế tiếp)
    TIMESTEPS = 12 
    X, y = [], []
    for i in range(len(data_scaled) - TIMESTEPS):
        X.append(data_scaled[i : i + TIMESTEPS])
        y.append(data_scaled[i + TIMESTEPS, 0]) # Mục tiêu dự đoán: 'flow'

    X, y = np.array(X), np.array(y)

    # Phân chia tập dữ liệu huấn luyện/kiểm thử theo trật tự thời gian (80/20)
    train_size = int(len(X) * 0.8)
    X_train, X_test = X[:train_size], X[train_size:]
    y_train, y_test = y[:train_size], y[train_size:]

    # Bộ thu phóng nghịch đảo riêng cho trường dữ liệu 'flow' để tính toán chỉ số thực tế
    target_scaler = MinMaxScaler()
    target_scaler.fit(data[:, 0].reshape(-1, 1))
    y_test_real = target_scaler.inverse_transform(y_test.reshape(-1, 1))

    print(f"--> Kích thước tập huấn luyện thực tế (Train): {X_train.shape}")
    print(f"--> Kích thước tập kiểm thử thực tế (Test)   : {X_test.shape}")

    # ==========================================
    # 2. XÂY DỰNG MÔ HÌNH VỚI DROPOUT 20%
    # ==========================================
    def build_network(model_name):
        model = Sequential()
        if model_name == 'SimpleRNN':
            model.add(SimpleRNN(64, input_shape=(TIMESTEPS, len(features)), activation='tanh', return_sequences=False))
        elif model_name == 'LSTM':
            model.add(LSTM(64, input_shape=(TIMESTEPS, len(features)), activation='tanh', return_sequences=False))
        elif model_name == 'GRU':
            model.add(GRU(64, input_shape=(TIMESTEPS, len(features)), activation='tanh', return_sequences=False))
        
        model.add(Dropout(0.20)) 
        model.add(Dense(1))
        model.compile(optimizer='adam', loss='mse')
        return model

    # Khởi tạo cấu trúc lưu trữ thông tin toán học
    history_logs = {}
    predictions_dict = {} # Lưu trữ kết quả dự báo để vẽ biểu đồ so sánh thực tế
    evaluation_metrics = []
    EPOCHS = 30
    BATCH_SIZE = 32

    # Thư mục lưu trữ đầu ra chuẩn trên Kaggle
    OUTPUT_DIR = "/kaggle/working/"
    os.makedirs(OUTPUT_DIR, exist_ok=True)

    # ==========================================
    # 3. HUẤN LUYỆN VÀ TRÍCH XUẤT CHỈ SỐ ĐÁNH GIÁ THỰC TẾ
    # ==========================================
    for m_label in ['SimpleRNN', 'LSTM', 'GRU']:
        print(f"\n======================================")
        print(f"👉 KHỞI CHẠY TIẾN TRÌNH HUẤN LUYỆN CHÍNH THỨC: {m_label}")
        print(f"======================================")
        
        net = build_network(m_label)
        
        # Tiến hành huấn luyện mô hình
        run_history = net.fit(
            X_train, y_train,
            epochs=EPOCHS,
            batch_size=BATCH_SIZE,
            validation_data=(X_test, y_test),
            verbose=1,
            callbacks=[EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)]
        )
        
        # Lưu lại nhật ký mất mát qua các Epoch
        history_logs[m_label] = run_history.history
        
        # Thực hiện dự đoán giá trị trên tập kiểm thử
        predictions_scaled = net.predict(X_test)
        # Chuyển đổi ngược về dải giá trị thực tế của bài toán gốc
        predictions_real = target_scaler.inverse_transform(predictions_scaled)
        predictions_dict[m_label] = predictions_real
        
        # 📊 TÍNH TOÁN 4 CHỈ SỐ HỌC THUẬT QUY CHUẨN
        mse = mean_squared_error(y_test_real, predictions_real)
        mae = mean_absolute_error(y_test_real, predictions_real)
        rmse = np.sqrt(mse)
        r2 = r2_score(y_test_real, predictions_real)
        
        # Lưu trữ vào danh sách tổng kết (Ẩn thông tin tỷ lệ mẫu)
        evaluation_metrics.append({
            "Mô hình": m_label,
            "Tập dữ liệu": "Toàn bộ dữ liệu gốc",
            "MAE": round(mae, 4),
            "MSE": round(mse, 4),
            "RMSE": round(rmse, 4),
            "R2 Score": round(r2, 4)
        })

    # ==========================================
    # 4. KHỞI TẠO BẢNG BIỂU VÀ CÁC BIỂU ĐỒ ĐỒ THỊ
    # ==========================================
    # Chuyển đổi kết quả sang định dạng bảng dữ liệu DataFrame
    summary_table = pd.DataFrame(evaluation_metrics)
    
    # Xuất file CSV báo cáo lưu trữ (Tên file chuẩn hóa không lộ 1/50)
    csv_report_path = os.path.join(OUTPUT_DIR, "model_evaluation_report.csv")
    summary_table.to_csv(csv_report_path, index=False)
    
    # Hiển thị bảng tổng hợp kết quả sạch
    print("\n" + "="*80)
    print(f"{'BẢNG TỔNG HỢP KẾT QUẢ SO SÁNH HIỆU NĂNG CÁC MÔ HÌNH':^80}")
    print("="*80)
    print(summary_table.to_string(index=False))
    print("="*80)
    print(f"[XÁC NHẬN] Đã xuất tệp dữ liệu số báo cáo (.csv) tại: {csv_report_path}\n")

    # BIỂU ĐỒ 1: Đồ thị đường loss phân tích tiến trình tối ưu hóa mạng
    plt.figure(figsize=(12, 6))
    graph_colors = {'SimpleRNN': '#E74C3C', 'LSTM': '#3498DB', 'GRU': '#2ECC71'}

    for m_label in history_logs:
        plt.plot(history_logs[m_label]['loss'], label=f'{m_label} - Train Loss', linestyle='--', color=graph_colors[m_label], alpha=0.6)
        plt.plot(history_logs[m_label]['val_loss'], label=f'{m_label} - Val Loss', linestyle='-', color=graph_colors[m_label], linewidth=2.5)

    plt.title('Đồ thị so sánh tiến trình mất mát (Loss Curve Analysis)', fontsize=14, fontweight='bold')
    plt.xlabel('Số lượng chu kỳ huấn luyện (Epochs)', fontsize=12)
    plt.ylabel('Sai số bình phương trung bình (MSE)', fontsize=12)
    plt.legend(loc='upper right', fontsize=10)
    plt.grid(True, linestyle=':', alpha=0.6)
    
    loss_chart_path = os.path.join(OUTPUT_DIR, "learning_curve_comparison.png")
    plt.savefig(loss_chart_path, dpi=300, bbox_inches='tight')
    plt.show()
    print(f"[XÁC NHẬN] Đã xuất file biểu đồ so sánh tiến trình học tại: {loss_chart_path}\n")

    # BIỂU ĐỒ 2 (BỔ SUNG CẤP THIẾT): Biểu đồ so sánh giá trị Thực tế vs Dự báo (Mô phỏng 150 bước thời gian cho dễ nhìn)
    plt.figure(figsize=(15, 6))
    PLOT_RANGE = 150 # Trích xuất 150 điểm dữ liệu liên tục để đồ thị không bị dính đặc vào nhau
    
    plt.plot(y_test_real[:PLOT_RANGE], label='Actual (Thực tế)', color='#2C3E50', linewidth=3, alpha=0.8)
    for m_label in predictions_dict:
        plt.plot(predictions_dict[m_label][:PLOT_RANGE], label=f'{m_label} Predicted (Dự báo)', color=graph_colors[m_label], linestyle='-', linewidth=1.8)

    plt.title('Đồ thị trực quan hóa kết quả dự báo lưu lượng dòng xe (Actual vs Predicted)', fontsize=14, fontweight='bold')
    plt.xlabel('Mốc thời gian kiểm thử (Timesteps)', fontsize=12)
    plt.ylabel('Giá trị lưu lượng (Traffic Flow)', fontsize=12)
    plt.legend(loc='upper right', fontsize=10)
    plt.grid(True, linestyle=':', alpha=0.6)
    
    predict_chart_path = os.path.join(OUTPUT_DIR, "prediction_comparison_analysis.png")
    plt.savefig(predict_chart_path, dpi=300, bbox_inches='tight')
    plt.show()
    print(f"[XÁC NHẬN] Đã xuất file biểu đồ so sánh giá trị dự báo thực tế tại: {predict_chart_path}")
```

```python
# ==========================================
# 5. LƯU MÔ HÌNH HUẤN LUYỆN
# ==========================================

SAVE_MODEL_DIR = "/kaggle/working/models/"
os.makedirs(SAVE_MODEL_DIR, exist_ok=True)

trained_models = {}

for m_label in ['SimpleRNN', 'LSTM', 'GRU']:

    print(f"\nĐang lưu mô hình: {m_label}")

    model = build_network(m_label)

    model.fit(
        X_train,
        y_train,
        epochs=EPOCHS,
        batch_size=BATCH_SIZE,
        validation_data=(X_test, y_test),
        verbose=0
    )

    model_save_path = os.path.join(
        SAVE_MODEL_DIR,
        f"{m_label.lower()}_traffic_model.h5"
    )

    model.save(model_save_path)

    trained_models[m_label] = model_save_path

    print(f"Đã lưu tại: {model_save_path}")

print("\nHoàn tất lưu toàn bộ mô hình.")
```
